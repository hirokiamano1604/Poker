```python
import pandas as pd
import scipy.stats
import numpy as np
import edhec_risk_kit as erk

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
me_m = pd.read_csv('../data/Portfolios_Formed_on_ME_monthly_EW.csv', header = 0, index_col = 0, parse_dates = True, na_values = -99.99)

rets = me_m[['Lo 20', 'Hi 20']]
rets.columns = ['SmallCap', 'LargeCap']
rets = rets/100
rets.index = pd.to_datetime(rets.index, format = '%Y%m')
rets.index = rets.index.to_period('M')
```

```python
# 1 What was the Annualized Return of the Lo 20 portfolio over the entire period?
wealth_index_Lo20 = 1000*(1+rets['SmallCap']).cumprod()
wealth_return_Lo20 = (wealth_index_Lo20.iloc[-1]-1000)/1000
numOfMonth_Lo20 = len(rets['SmallCap'])
numOfYear_Lo20 = numOfMonth_Lo20/12
AnnualReturn_Lo20 = wealth_return_Lo20 **(1/numOfYear_Lo20)-1
AnnualReturn_Lo20
```

```python
# 2 What was the Annualized Volatility of the Lo 20 portfolio over the entire period? 
monthly_vol_Lo20= rets['SmallCap'].std(ddof=0)
annual_vol_Lo20 = monthly_vol_Lo20 * np.sqrt(12)
annual_vol_Lo20
```

```python
# 3 What was the Annualized Return of the Hi 20 portfolio over the entire period?     
wealth_index_Hi20 = 1000*(1+rets['LargeCap']).cumprod()
wealth_return_Hi20 = (wealth_index_Hi20.iloc[-1]-1000)/1000
numOfMonth_Hi20 = len(rets['LargeCap'])
numOfYear_Hi20 = numOfMonth_Hi20/12
AnnualReturn_Hi20 = wealth_return_Hi20**(1/numOfYear_Hi20)-1
AnnualReturn_Hi20
```

```python
# 4 What was the Annualized Volatility of the Hi 20 portfolio over the entire period ?  
monthly_vol_Hi20= rets['LargeCap'].std(ddof=0)
annual_vol_Hi20 = monthly_vol_Hi20 * np.sqrt(12)
annual_vol_Hi20
```

```python
# 5 What was the Annualized Return of the Lo 20 portfolio over the period 1999 - 2015 (both inclusive)? 
rets_99_15 = rets.loc["1999":"2015"]
wealth_index_Lo20_9915 = 1000*(1+rets_99_15['SmallCap']).cumprod()
wealth_return_Lo20_9915 = (wealth_index_Lo20_9915.iloc[-1]-1000)/1000
numOfMonth_Lo20_9915 = len(rets_99_15['SmallCap'])
numOfYear_Lo20_9915 = numOfMonth_Lo20_9915/12
AnnualReturn_Lo20_9915 = wealth_return_Lo20_9915**(1/numOfYear_Lo20_9915)-1
AnnualReturn_Lo20_9915
```

```python
# 6 What was the Annualized Volatility of the Lo 20 portfolio over the period 1999 - 2015 (both inclusive)?
monthly_vol_Lo20_9915= rets_99_15['SmallCap'].std(ddof=0)
annual_vol_Lo20_9915 = monthly_vol_Lo20_9915 * np.sqrt(12)
annual_vol_Lo20_9915
```

```python
# 7 What was the Annualized Return of the Hi 20 portfolio over the period  1999 - 2015 (both inclusive)?
wealth_index_Hi20_9915 = 1000*(1+rets_99_15['LargeCap']).cumprod()
wealth_return_Hi20_9915 = (wealth_index_Hi20_9915.iloc[-1]-1000)/1000
numOfMonth_Hi20_9915 = len(rets_99_15['LargeCap'])
numOfYear_Hi20_9915 = numOfMonth_Hi20_9915/12
AnnualReturn_Hi20_9915 = wealth_return_Hi20_9915**(1/numOfYear_Hi20_9915)-1
AnnualReturn_Hi20_9915
```

```python
# 8 What was the Annualized Volatility of the Hi 20 portfolio over the period 1999 - 2015 (both inclusive)?  
monthly_vol_Hi20_9915= rets_99_15['LargeCap'].std(ddof=0)
annual_vol_Hi20_9915 = monthly_vol_Hi20_9915 * np.sqrt(12)
annual_vol_Hi20_9915
```

```python
# 9 What was the Max Drawdown (expressed as a positive number) experienced over the 1999-2015 period in the SmallCap (Lo 20) portfolio?
previous_peak_Lo20_9915 = wealth_index_Lo20_9915.cummax()
drawdown_Lo20_9915 = (wealth_index_Lo20_9915 - previous_peak_Lo20_9915)/previous_peak_Lo20_9915
-drawdown_Lo20_9915.min()
```

```python
# 10 At the end of which month over the period 1999-2015 did that maximum drawdown on the SmallCap (Lo 20) portfolio occur? 
drawdown_Lo20_9915.idxmin()
```

```python
# 11 What was the Max Drawdown (expressed as a positive number) experienced over the 1999-2015 period in the LargeCap (Hi 20) portfolio?
previous_peak_Hi20_9915 = wealth_index_Hi20_9915.cummax()
drawdown_Hi20_9915 = (wealth_index_Hi20_9915 - previous_peak_Hi20_9915)/previous_peak_Hi20_9915
-drawdown_Hi20_9915.min()
```

```python
#12 Over the period 1999-2015, at the end of which month did that maximum drawdown of the LargeCap (Hi 20) portfolio occur? 
drawdown_Hi20_9915.idxmin()
```
