```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
```

```python
ind = erk.get_ind_returns()
er = erk.annualize_rets(ind["2013":"2017"],12)
cov = ind["2013":"2017"].cov()
```

```python
#5
l = ["Books","Steel","Oil","Mines"]
msrWeights = erk.msr(0.1, er[l], cov.loc[l,l])
msrWeights
```

```python
#8
gmvWeights = erk.gmv(cov.loc[l,l])
gmvWeights
```

```python
#11
er_2018 = ind.loc["2018"][l]
cov_2018 = er_2018.cov() # type: ignore
msrVol = erk.portfolio_vol(msrWeights, cov_2018)
annualMsrVol = msrVol * np.sqrt(12)
annualMsrVol
```

```python
gmvVol = erk.portfolio_vol(gmvWeights, cov_2018)
annualGmvVol = gmvVol * np.sqrt(12)
annualGmvVol
```
